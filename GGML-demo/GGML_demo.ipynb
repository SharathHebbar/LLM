```python
MODEL_NAME = "TheBloke/Llama-2-13B-chat-GGML"
config = {
    'max_new_tokens': 1024,

    'temperature': 0.9,
    'stream': True
}
```

```python
from ctransformers import AutoModelForCausalLM

MODEL_NAME = "TheBloke/Llama-2-13B-chat-GGML"
config = {
    'max_new_tokens': 1024,

    'temperature': 0.9,
    'stream': True
}
```

```python
llm = AutoModelForCausalLM.from_pretrained(
    MODEL_NAME,
    gpu_layers=130 #110 for 7b, 130 for 13b
    **config
    )
```

```python
llm("Hii What are you doing", stream=False)
```
